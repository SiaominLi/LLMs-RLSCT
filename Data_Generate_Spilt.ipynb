## 生成數據集

```python
import json
import random

class CFG:
    def __init__(self, rules):
        self.rules = rules

    def generate_tree(self, symbol):
        if symbol not in self.rules:
            return symbol, []  # Return an empty list if no children
        else:
            rhs = random.choice(self.rules[symbol].split('|'))
            children = [self.generate_tree(sym) for sym in rhs.split()]
            return symbol, children

    def print_tree(self, tree, depth=0):
        symbol, children = tree
        print("  " * depth + symbol)
        for child in children:
            self.print_tree(child, depth + 1)

    def flatten_tree(self, tree):
        symbol, children = tree
        if children:
            return ' '.join(self.flatten_tree(child) for child in children)
        else:
            return symbol

    def exchange_children(self, tree):
        symbol, children = tree
        if len(children) == 2:
            # Swap the children
            children[0], children[1] = children[1], children[0]
        for child in children:
            self.exchange_children(child)
        return tree

# 讀取 JSON 檔案
with open('/content/grammar_rule_correct.json', 'r') as file:
    rules = json.load(file)

# 創建語法對象
grammar = CFG(rules)

# 生成 ??? 對句子
output_pairs = {}
for _ in range(1005):
    original_syntax_tree = grammar.generate_tree('S')
    s1 = grammar.flatten_tree(original_syntax_tree)

    # Exchanging children and printing the modified syntax tree
    modified_syntax_tree = grammar.exchange_children(original_syntax_tree)
    s2 = grammar.flatten_tree(modified_syntax_tree)
    # print(f"Original: {s1}\nModified: {s2}\n")
    # output_pairs.append([s1,s2])
    output_pairs[s1] = s2

print(len(output_pairs))

output_list = []
for s1, s2 in output_pairs.items():
    output_list.append({s1: s2})

# 將輸出保存為 JSON 檔案
output_json = json.dumps(output_list, indent=2)
with open('/content/eng_sentences1000.json', 'w') as output_file:
    output_file.write(output_json)
```

## 切割數據集

```python

import json
import random

# 讀取原始JSON檔案
with open('eng_sentences1000_長度為20.json', 'r') as f:
    data = json.load(f)

# 隨機選擇100筆資料
random_data = random.sample(data, 100)

# 寫入新的JSON檔案
with open('eng_sentences_cut100_長度為20.json', 'w') as f:
    json.dump(random_data, f, indent=2)

# 將已經隨機選擇的100筆資料從原始資料中移除
remaining_data = [d for d in data if d not in random_data]

# 將剩下的990筆資料存儲到一個新的JSON檔案中
with open('eng_sentences_cut900_長度為20.json', 'w') as f:
    json.dump(remaining_data, f, indent=2)
```
