### 詞性retrieve

```python
# 初始化記錄相似度的列表
average_similarities = []

#如果用dic存的話，有可能第二個query也用到一樣的句子，就沒有存到dic裡了
retrieve_output = []
for eng in eng_data:
    sentence = list(eng.keys())[0]
    # results = {}
    # 給定的句子
    print(f'query：{sentence}')
    # results[sentence] = item["調整後句子"]
    retrieve_output.append([list(eng.keys())[0],list(eng.values())[0]])

    # 將句子分解成單詞並找到每個單詞的詞性
    pos_sequence = []
    for word in sentence.split():
        if word in data:
            pos_sequence.append(data[word])
        else:
            # 如果單詞在字典中找不到對應的詞性，可以將其標記為未知
            pos_sequence.append("Unknown")

    # 輸出詞性排列
    pos_sequence_str = ' '.join(pos_sequence)
    print("詞性排列：", pos_sequence_str)

    # 初始化最接近的10句
    closest_sentences = []
    max_closest_sentences = 10

    # 遍歷每一對句子，計算其詞性排列與給定排列的相似度
    for pair in paired_data:
        # print(pair)
        original_sentence = list(pair.keys())[0]
        reordered_sentence = list(pair.values())[0]

        ori_sequence = []
        for ori_word in original_sentence.split():
            if ori_word in data:
                ori_sequence.append(data[ori_word])
            else:
                # 如果單詞在字典中找不到對應的詞性，可以將其標記為未知
                ori_sequence.append("Unknown")

        ori_sequence_str = ' '.join(ori_sequence)

        # 假設句子的詞性排列是以空格分隔的字串，可以將其分割成單詞列表
        ori_pos_sequence = ori_sequence_str.split()
        pair_pos_sequence = pos_sequence_str.split()

        if len(pair_pos_sequence) < len(ori_pos_sequence):
            continue

        # 計算兩個詞性排列的相似度
        similarity = sum(1 for a, b in zip(ori_pos_sequence, pair_pos_sequence) if a == b) / len(pair_pos_sequence)

        # 如果相似度比目前最接近的10句中的最小相似度大，則將這句加入最接近的10句中
        if len(closest_sentences) < max_closest_sentences or similarity > closest_sentences[-1][1]:
            closest_sentences.append((original_sentence, similarity))
            closest_sentences.sort(key=lambda x: x[1], reverse=True)
            closest_sentences = closest_sentences[:max_closest_sentences]

    # 計算最接近的10句的相似度平均值
    avg_similarity = sum(similarity for _, similarity in closest_sentences) / len(closest_sentences)
    average_similarities.append(avg_similarity)

    # 輸出最接近的10句
    # print("\n最接近的10句：")
    # for sentence, similarity in closest_sentences:
    #     print(f"{sentence} - 相似度: {similarity}")
    paired_data_dict = {}
    for item in paired_data:
        key = list(item.keys())[0]
        value = list(item.values())[0]
        paired_data_dict[key] = value
    # print(paired_data_dict["have he from they you"])

    for sentence, similarity in closest_sentences:
        original_value = sentence
        reordered_value = paired_data_dict[sentence]
        print(f"{original_value}\n{reordered_value}- 相似度: {similarity}\n")
        retrieve_output.append([original_value,reordered_value])
        # results[original_value] = reordered_value
    print("------------------------------------------------------------")

# 輸出所有100組的相似度平均值
for idx, avg_sim in enumerate(average_similarities):
    print(f"第 {idx+1} 組的相似度平均值：{avg_sim}")

# 也可以計算所有組的平均相似度
overall_avg_similarity = sum(average_similarities) / len(average_similarities)
print(f"\n所有組的相似度平均值：{overall_avg_similarity}")

print(f"Output句子對數 ： {len(retrieve_output)}")
# output_json = json.dumps(output, indent=2)
# with open('/content/eng範例10_POS_retrieval.json', 'w') as output_file:
#     output_file.write(output_json)
```

### 凱薩加密

```python
import json

def caesar_cipher(text, shift):
    result = ''
    for char in text:
        if char.isalpha():
            ascii_offset = ord('a') if char.islower() else ord('A')
            result += chr((ord(char) - ascii_offset + shift) % 26 + ascii_offset)
        else:
            result += char
    return result

# 對每一個句子進行凱薩加密
caesar_output = []
for pair in retrieve_output:
    encry1 = caesar_cipher(pair[0], shift=10)
    encry2 = caesar_cipher(pair[1], shift=10)
    caesar_output.append([encry1,encry2])
print(f"Output句子對數 ： {len(caesar_output)}")
```

```python
import json
from nltk.translate.bleu_score import sentence_bleu
from string import punctuation

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', punctuation))

with open("encrypted100wPOS_長度為16_gpt35.json", "r", encoding="utf-8") as file:
    pair_data = json.load(file)

# Calculate BLEU scores and save results
bs1_avg = []
bs2_avg = []
bs3_avg = []
bs4_avg = []
results = []

for sen in pair_data:
    origin = sen[0]
    modified = sen[1]
    guess = sen[2]

    m = [remove_punctuation(modified).lower().split()]
    g = remove_punctuation(guess).lower().split()

    #print(m + g)

    bleu1 = sentence_bleu(m, g, weights=(1, 0, 0, 0))
    bleu2 = sentence_bleu(m, g, weights=(0.5, 0.5, 0, 0))
    bleu3 = sentence_bleu(m, g, weights=(0.33, 0.33, 0.33, 0))
    bleu4 = sentence_bleu(m, g, weights=(0.25, 0.25, 0.25, 0.25))

    num_list = [str(bleu1), str(bleu2), str(bleu3), str(bleu4)]
    bleu = ' , '.join(num_list)

    bs1_avg.append(bleu1)  # BLEU-1 score
    bs2_avg.append(bleu2)  # BLEU-2 score
    bs3_avg.append(bleu3)  # BLEU-3 score
    bs4_avg.append(bleu4)  # BLEU-4 score
    #bleu_scores_dict[str(sen)] = {'BLEU-1': bleu1, 'BLEU-2': bleu2, 'BLEU-3': bleu3, 'BLEU-4': bleu4}

    results.append({
        "原句": origin,
        "調整後句子": modified,
        "GPT回應": guess,
        "BLEU 分數": bleu
    })

avg_bleu_score1 = sum(bs1_avg) / len(bs1_avg)
avg_bleu_score2 = sum(bs2_avg) / len(bs2_avg)
avg_bleu_score3 = sum(bs3_avg) / len(bs3_avg)
avg_bleu_score4 = sum(bs4_avg) / len(bs4_avg)

results.append({
    "平均 BLEU1 分數": avg_bleu_score1,
    "平均 BLEU2 分數": avg_bleu_score2,
    "平均 BLEU3 分數": avg_bleu_score3,
    "平均 BLEU4 分數": avg_bleu_score4
})

# Save BLEU scores and results
with open("/content/encrypted100wPOS_長度為16_gpt35_bleusc.json", "w", encoding="utf-8") as output_file:
    json.dump(results, output_file, ensure_ascii=False, indent=4)
```
